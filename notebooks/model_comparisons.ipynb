### Compare Training Distribution

```python
def training_data_dist(class_count: dict,
                       title: str):
    total = sum(class_count.values())
    class_dist = {}
    for key, val in class_count.items():
        class_dist[key] = round((val/total)*100,1)
    
    df = pd.DataFrame({'count': class_count, 'distribution': class_dist})
    fig, ax = plt.subplots()
    wedges, texts, autotexts = ax.pie(df['count'], 
                                      labels=df.index, 
                                      autopct='%1.1f%%', 
                                      textprops=dict(color="black"))
    
    # Annotate with the count in addition to the percentage
    for i, text in enumerate(autotexts):
        text.set_text(f"{text.get_text()}\n({df['count'].iloc[i]})")
    
    plt.title(title)
    plt.show()
```

```python
training_data_dist(class_count = {0.0: 42755, 1.0: 36369, 2.0: 76856, 3.0: 38256},
                   title= 'v26 Class Distribution')
```
