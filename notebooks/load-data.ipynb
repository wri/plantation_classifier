```python
import numpy as np
import hickle as hkl
import os

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def load_slope(idx,
               directory = 'train-slope/'):
    """
    Slope is stored as a 32 x 32 float32 array with border information
    Needs to be converted to 14 x 14 to match the labels
    """
    x = np.load(directory + str(idx) + '.npy').squeeze()
    original_shape = x.shape
    
    border_x = (x.shape[0] - 14) // 2
    border_y = (x.shape[1] - 14) // 2
    slope = x[border_x:-border_x, border_y:-border_y]
    slope = slope[np.newaxis]
    
    print(f'{idx} slope: {original_shape} -> {slope.shape}, {slope.dtype}')
    return slope


def load_s1(idx,
            directory = 'train-s1/'):
    """
    S1 is stored as a (12, 32, 32, 2) array
    Convert the monthly mosaics to an annual median, and remove the border
    To convert to a (14 14, 2) array
    """
    
    s1 = hkl.load(directory + str(idx) + '.hkl')
    original_shape = s1.shape
    
    if len(s1.shape) == 4:
        s1 = np.median(s1, axis = 0)
        
    border_x = (s1.shape[0] - 14) // 2
    border_y = (s1.shape[1] - 14) // 2
    s1 = s1[border_x:-border_x, border_y:-border_y]
    s1 = s1.astype(np.float32)
    
    print(f'{idx} s1: {original_shape} -> {s1.shape}, {s1.dtype}')
    return s1.astype(np.float32)


def load_s2(idx,
            directory = 'train-s2/'):
    
    """
    S2 is stored as a (12, 28, 28, 11) array. 
    The last axis index is the date of the imagery in the first axis. Remove.
    The monthly images need to be converted to an annual median.
    Remove the border to correspond to the labels.
    The S2 is stored as uint16, convert to float32
    """
    s2 = hkl.load(directory + str(idx) + '.hkl')
    original_shape = s2.shape
    
    if s2.shape[-1] == 11:
        s2 = np.delete(s2, -1, -1)
    
    if len(s2.shape) == 4:
        s2 = np.median(s2, axis = 0)
        
    if not isinstance(s2.flat[0], np.floating):
        assert np.max(s2) > 1
        s2 = s2.astype(np.float32) / 65535
        assert np.max(s2) < 1
        
    border_x = (s2.shape[0] - 14) // 2
    border_y = (s2.shape[1] - 14) // 2
    s2 = s2[border_x:-border_x, border_y:-border_y].astype(np.float32)
    print(f'{idx} s2: {original_shape} -> {s2.shape}, {s2.dtype}')
    return s2


def load_feats(idx,
            directory = 'train-features/'):
    '''
    The features are stored as a (14, 14, 65) array.
    The last axis are the feature dimensions. 
    Index 0 is the tree cover prediction from the full TML model
    Index 1 - 33 are high level features
    Index 33 - 65 are low level features
    '''
    feats = hkl.load(directory + str(idx) + '.hkl').astype(np.float32)
    print(f'{idx} feats: {feats.shape}, {feats.dtype}')
    return feats

def load_label(idx,
            directory = 'train-labels/'):
    '''
    The  labels are stored as a binary 14, 14 array
    '''
    labels = np.load(directory + str(idx) + '.npy')[np.newaxis]
    print(f'{idx} labels: {labels.shape}, {labels.dtype}')
    return labels
```

```python
a = load_slope(0)
b = load_s1(0)
c = load_s2(0)
d = load_feats(0)
e = load_label(0)
```

```python
def make_x_sample(slope, s1, s2, feats):
    n_feats = 1 + s1.shape[-1] + s2.shape[-1] + feats.shape[-1]
    sample = np.empty((14, 14, n_feats))
    sample[..., 0] = slope
    sample[..., 1:3] = s1
    sample[..., 3:13] = s2
    sample[..., 13:] = feats
    return sample
```

```python
f = make_x_sample(a, b, c, d)
```

```python
def multiplot(matrices, nrows = 13, ncols = 6):
    '''Docstring
    
         Parameters:
          matrices (list):
          nrows (int):
          
         Returns:
          None
    '''
    fig, axs = plt.subplots(ncols = ncols, nrows = nrows)
    fig.set_size_inches(18, 3.25*nrows)
    to_iter = [[x for x in range(i, i + ncols + 1)] for i in range(0, nrows*ncols, ncols)]
    counter = 0
    for r in range(1, nrows + 1):
        min_i = min(to_iter[r-1])
        max_i = max(to_iter[r-1])
        for i in range(ncols):
            sns.heatmap(data = matrices[..., counter], ax = axs[r - 1, i], cbar = False,
                       cmap = sns.color_palette("viridis", as_cmap=True))
            axs[r - 1, i].set_xlabel("")
            axs[r - 1, i].set_ylabel("")
            axs[r - 1, i].set_yticks([])
            axs[r - 1, i].set_xticks([])
            counter += 1
    plt.show
```

```python
multiplot(f)
```
